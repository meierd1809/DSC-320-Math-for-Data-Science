## Plotting Linear Models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from statistics import mean

car_data = pd.read_csv('car_data.csv')
```

# Exercise 1

```python
# Getting the values for both the car's weight and the car's highway miles per gallon

car_weight = car_data['weight'].values
car_hwy_mpg = car_data['hwy_mpg'].values

# Applying the car weight to the x value and car highway miles per gallon to the y value

x = car_weight
y = car_hwy_mpg

# Creating the initial figure 
plt.rcParams["figure.figsize"] = [10, 5]

# Creating the scatterplot based on the x and y values and making the points blue
plt.scatter(x, y, color ='blue')

# Adding a title to the plot, and a label for the x and y axis
plt.title('Highway MPG vs. Weight of Car (lbs)')
plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles Per Gallon');
```

# Exercise 2

Based on the plot above, it would appear that the more the car weighs, the less highway miles per gallon the car gets.

# Exercise 3

If I were to build a linear model using this data, I would expect the slope to be negative. The reason for this is that there is a negative correlation in the data. As the X (Car Weight) increases, the Y (Highway MPG) decreases, which would result in a negative slope.

# Exercise 4

Having a slope of -0.05 means first that as the weight of the car increases, the highway mpg decreases. Given that the slope is very small at -0.05, the change in y as x changes is very minimal.

# Exercise 5

```python
# Creating the initial figure
plt.rcParams["figure.figsize"] = [10, 5]

# Plotting the x and y values as a scatter plot and making the points blue
plt.scatter(x, y, color ='blue')

# Adding a title to the plot and labeling the x and y axis
plt.title('Highway MPG vs. Weight of Car (lbs)')
plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles Per Gallon')

# Adding a line of best fit
plt.plot([500, 4000], [47.5, 20], 'k--')

#Determining the slope based on the values used in creating the line of best fit
slope = '{:.2f}'.format((15-47.5)/(4000-500))
# This takes the slope, multiplies it by an x value, and then adds that to the subsequent y value.
y_intercept = 47.5 + 0.01 * 500

#This is plotting the equation of the best fit line
plt.text(3000, 50, f"y = {slope}x + {y_intercept}", 
         bbox=dict(facecolor='yellow', alpha=1));

# Slope = -0.01 y-intercept = 52.5 based on this line of best fit
```

# Exercise 6

```python
from sklearn.linear_model import LinearRegression

# This is adding the car weight values to the car_weight variable and adjusting it so it will work with the 
# Linear Regression model, along with adding the car highway miles per gallon to the hcar_hwy_mpg variable
X = car_data['weight'].values[:,np.newaxis]
Y = car_data['hwy_mpg'].values

# Creating the Linear Regression model and fitting the x and y values
model = LinearRegression().fit(X, Y)

y_pred = model.predict(X)

#Finding the slope and y_intercept of the data
slope = model.coef_
slope = slope[0]
y_intercept = '{:.2f}'.format(model.intercept_)

#Plotting the x and y values as a scatterplot and making the points green
plt.scatter(X, Y, color='g')

#Plotting the actual best fit line
plt.plot(X, y_pred ,color='r')

plt.text(3000, 50, f"y = {'{:.2f}'.format(slope)}x + {y_intercept}", 
         bbox=dict(facecolor='yellow', alpha=1));

#Adding a title to the plot, along with a label for the x and y axis
plt.title('Highway MPG vs. Weight of Car (lbs)')
plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles Per Gallon');
```

# Exercise 7

```python
from sklearn.metrics import mean_squared_error as mse

print('Root Mean Square Error of Prediction Line = ' + '{:.2f}'.format(mse(y, y_pred)))
print('Root Mean Square Error of Actual Best Fit Line = ' + '{:.2f}'.format(mse(Y, y_pred)))
```
