# 1. Solving a System of Linear Equations

```python
import numpy as np
```

```python
# First step is to create a matrix for the numbers on the left side of the equation, and a separate matrix for the answer on 
# the right side of the equation

Matrix_A = np.matrix([[27, -10, 4, -29], [-16, 5, -2, 18], [-17, 4, -2, 20], [-7, 2, -1, 8]])
Matrix_B = np.matrix([[1], [-1], [0], [1]])
```

```python
# Next step is to find the inverse of Matrix_A

Matrix_A_inv = np.linalg.inv(Matrix_A)
```

```python
# To find the x values, you take the inverse of Matrix A and multiply it by Matrix B

Matrix_X = Matrix_A_inv * Matrix_B

print("X_1 =", Matrix_X[0]) 
print("X_2 =", Matrix_X[1]) 
print("X_3 =", Matrix_X[2])
print("X_4 =", Matrix_X[3])
```

```python
# Gets the same answer, but requires less code

np.linalg.solve(Matrix_A, Matrix_B)
```

# 2. Visualizing a Linear Transformation

## (a)

```python
def T(v):
    A = np.matrix([[np.sqrt(2)/2, -(np.sqrt(2)/2)], [np.sqrt(2)/2, np.sqrt(2)/2]])
    return A @ v
```

```python
X = np.array([2, -2])
```

```python
T(X)
```

## (b)

```python
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)

xx, yy = np.meshgrid(x, y)
```

```python
plt.scatter(xx, yy, s=20, c=xx+yy)
plt.xlabel("xx")
plt.ylabel("yy");
```

```python
T = np.array([
[-1, 0],
[0, -1]
])
```

```python
xy = np.vstack([xx.flatten(), yy.flatten()])
xy.shape
```

```python
trans = T @ xy
trans.shape
```

```python
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)
```

```python
f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

```python
T = np.array([
[1.3, -2.4],
[0.1, 2]
])
trans = T @ xy
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)
f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```
