# 2. Principal Component Analysis

### (a)

```python
import pandas as pd
import matplotlib.pyplot as plt

video_games = pd.read_csv("video_game_data.csv")
video_games.plot.scatter(x = "user_score", y = 'critic_score');
```

### (b)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca.fit(video_games[['user_score', 'critic_score']])

components = pca.components_
```

```python
ax = video_games.plot.scatter(x = 'user_score', y = 'critic_score')

ax.arrow(0, 0, components[0][0], components[0][1], fc = 'r', ec='r')
ax.arrow(0, 0, components[1][0], components[1][1], fc = 'g', ec='g');
```
