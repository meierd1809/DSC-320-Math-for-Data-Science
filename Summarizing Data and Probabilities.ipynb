# 1. Summarizing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
qb_stats = pd.read_csv("qb_stats.csv")
qb_stats
```

## a)

```python
# Using numeric_only to avoid trying to pull any columns that have non-numeric data

qb_stats.mean(numeric_only=True)
```

## b)

```python
# Using numeric_only to avoid trying to pull any columns that have non-numeric data

qb_stats.std(numeric_only=True)
```

## c)

```python
# Decided on 10 bins to see more distribution of the data

plt.hist(qb_stats[['yds']], bins = 10, color="green")
plt.ylabel('Count of QBs')
plt.xlabel('Total Yards')
plt.title("Distribution of QBs based on Total Yards")
```

## d)

```python
plt.boxplot(qb_stats[['td']])
plt.ylabel('Total Touchdowns')
```

```python
# The following code is to help determine what the outliers are in the data

qb_td_arr = np.array(qb_stats['td'])
# finding the 1st quartile
q1 = np.quantile(qb_td_arr, 0.25)
 
# finding the 3rd quartile
q3 = np.quantile(qb_td_arr, 0.75)
med = np.median(qb_td_arr)
 
# finding the iqr region
iqr = q3-q1
 
# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

outliers = qb_td_arr[(qb_td_arr <= lower_bound) | (qb_td_arr >= upper_bound)]
print("The outliers in this boxplot are:", outliers)
```

```python
# This returns the name of the quarterback(s) who were associated with the outliers in touchdown numbers

outlier_qbs = qb_stats.query(f"td == {outliers}")["name"]
print("The outlier QBs in this data are:", outlier_qbs)
```

# 2. Calculating Probabilities from Data

## a)

```python
survey_df = pd.read_csv("survey_data.csv")
survey_df.head(15)
```

```python
# Total survey data

total_data = len(survey_df)

# This is filtering out all of the rows where the hair color is brown

brown_hair = len(survey_df[survey_df['hair_color'] == 'brown'])
```

```python
# This is taking the count of all of the rows of brown hair and dividing it by the count of all of the rows

brown_hair_prob = "{0:.2f}".format(brown_hair / total_data * 100) + "%"
print("The probability that a college student as brown hair is", brown_hair_prob)
```

## b)

```python
# This filters all of the rows where eye color is blue

blue_eyes = len(survey_df[survey_df['eye_color'] == "blue"])

# This is taking the count of all of the rows of blue eyes and dividing it by the count of all of the rows

blue_eyes_prob = "{0:.2f}".format(blue_eyes / total_data * 100) + "%"
print("The probability that a college student as blue eyes is", blue_eyes_prob)
```

## c)

```python
# This is filtering the rows of all of those with brown hair that happen to have blue eyes

brown_hair_blue_eyes = len(survey_df[(survey_df['hair_color'] == "brown") & (survey_df['eye_color'] == "blue")])

# This is taking the count of those with brown hair and blue eyes and dividing it by the count of everyone with brown hair,
# Regardless of eye color

brown_hair_blue_eyes_prob = "{0:.2f}".format(brown_hair_blue_eyes / brown_hair * 100)   + "%"
```

```python
print("The probability that a college student as blue eyes given that they have brown hair is", 
     brown_hair_blue_eyes_prob) 
```

## d)

```python
# This is taking the count of the previous set of data and dividing it by the count of everyone with blue eyes, regardless of 
# Hair color

blue_eyes_brown_hair_prob = "{0:.2f}".format(brown_hair_blue_eyes/blue_eyes * 100) + "%"
```

```python
print("The probability that a college student has brown hair given that they have blue eyes is", 
     blue_eyes_brown_hair_prob) 
```

## e)

```python
print("Brown hair:", brown_hair_prob)
print("Brown hair given blue eyes:", blue_eyes_brown_hair_prob)
print()
print("Blue eyes:", blue_eyes_prob)
print("Blue eyes given brown hair:", brown_hair_blue_eyes_prob)
```

Based on the results above, I would say that having brown hair and blue eyes are dependent on each other. The reason I say this is that, for example, the probability of having brown hair changes when you introduce the condition of the eyes being blue because it decreases by roughly 6%. At the same time, the probability of having blue eyes decreases by 3% when the condition of having brown hair is introduced. If the probability of having brown hair stayed the same regardless of eye color (and vice versa if hair color changed while eye color stayed the same), then they would be independent of each other.
